```python
'''
13. Elabore um algoritmo que leia um número de entrada que indicará a quantidade de
registros a serem lidos (N). Em seguida algoritmo deve solicitar o sexo (M/F) e idade de N
pessoas e ao final apresentar a média de idade de ambos os gêneros catalogados.'''

quantidade = int(input("Informe o número de registros: "))
pessoas = []

i = 1
while True:
  sexo = input("Informe o sexo: ")
  idade = int(input("Informe a idade:"))

  pessoas.append({"sexo": sexo, "idade": idade})

  quantidade -= 1
  i += 1

  if quantidade < 1:
    break

if pessoas:
  feminino = sum(1 for pessoa in pessoas if pessoa.get("sexo") == "F")
  soma_idade_feminino = sum(pessoa.get("idade") for pessoa in pessoas if pessoa.get("sexo") == "F")
  media_idade_feminino = soma_idade_feminino / feminino if feminino else 0
  print(f"Total de mulheres é {feminino}. Com uma média de idade de {media_idade_feminino}")
if pessoas:
   masculino = sum(1 for pessoa in pessoas if pessoa.get("sexo") == "M")
   soma_idade_masculino = sum(pessoa.get("idade") for pessoa in pessoas if pessoa.get("sexo") == "M")
   media_idade_masculino = soma_idade_masculino / masculino if masculino else 0
   print(f"Total de homens é {masculino}. Com uma média de idade de {media_idade_masculino:.1f}")

```

```python
'''
16. Escreva um programa que vá solicitando as idades dos alunos da sala até que todos
sejam informados (perguntar ao usuário se deseja informar a idade do próximo aluno). Ao
final apresentar a idade do mais novo, a idade do mais velho, Quantos alunos têm mais de
18 anos, quantos alunos têm até 18 anos, a média aritmética e a mediana.'''

idades = []

while True:
  idade = int(input("informa a idade: "))
  idades.append(idade)

  continuar = input("Deseja informar a idade do próximo aluno? ")

  if continuar.lower() != "sim":
    break


mais_velha = max(idades)
mais_nova = min(idades)

maiores_18 = sum(1 for idade in idades if idade > 18)
ate_18 = sum(1 for idade in idades if idade <= 18)

#médias
soma_idades = sum(idades)
total_alunos = len(idades)
media_aritmetica = soma_idades/total_alunos

#Mediana
idades_ordenadas = sorted(idades)
tamanho = len(idades_ordenadas)

if tamanho % 2 == 1:
  # Se ímpar, pega o valor central
  mediana = idades_ordenadas[tamanho // 2]
else:
   # Se par, média dos dois centrais
  indice_1 = tamanho // 2 - 1
  indice_2 = tamanho // 2
  mediana = (idades_ordenadas[indice_1] + idades_ordenadas[indice_2]) / 2

print(f"Idade do mais novo: {mais_nova}")
print(f"Idade do mais velho: {mais_velha}")
print("---------------------------------")
print(f"Alunos com mais de 18 anos: {maiores_18}")
print(f"Alunos com até 18 anos: {ate_18}")
print("---------------------------------")
print(f"Média aritmética: {media_aritmetica:.2f}")
print(f"Mediana: {mediana:.2f}")
```
